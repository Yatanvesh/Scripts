```python
import xmltodict
import json
import fnmatch
import os, sys
import sumolib
if 'SUMO_HOME' in os.environ:
    tools = os.path.join(os.environ['SUMO_HOME'], 'tools')
    sys.path.append(tools)
else:
    sys.exit("please declare environment variable 'SUMO_HOME'")
```

```python
with open('fcd.xml') as fd:
    fcdFile = xmltodict.parse(fd.read())

netFile = ''
for file in os.listdir('.'):
    if fnmatch.fnmatch(file, '*.net.xml'):
        netFile = file
        
net = sumolib.net.readNet(netFile)
```

```python
timestamps = fcdFile['fcd-export']['timestep']
```

```python
output = []

for i in range(0, len(timestamps), 2):
    if i>500:
        break
    timestamp = timestamps[i]
    vehicles = []
    if isinstance( timestamp['vehicle'], list ):
        for vehicle in timestamp['vehicle']:
#             lng, lat = net.convertXY2LonLat(float(vehicle['@y']), float(vehicle['@x']) )
            obj = {"id": vehicle['@id'], "lng":vehicle['@x'], 'lat':vehicle['@y'], 'angle':vehicle['@angle'], 'type':vehicle['@type']}
            vehicles.append(obj)
    else:
        vehicle = timestamp['vehicle']
#         lng, lat = net.convertXY2LonLat(float(), float())
        obj = {"id": vehicle['@id'], "lng":vehicle['@x'], 'lat':vehicle['@y'], 'angle':vehicle['@angle'], 'type':vehicle['@type']}
        vehicles.append(obj)
        
    info = {'time':timestamp['@time'],'vehicles': vehicles}
    output.append(info)
```

```python
with open('fcd_output.json', 'w', encoding='utf-8') as f:
    json.dump(output, f, ensure_ascii=False, indent=4)
```

```python
# lightData = [{
#     'x':'732.41',
#     'y':'608.63'
# },
# {
#     'x':'1105.17',
#     'y':"861.78"
# },
# {
#     'x':'1094.11',
#     'y':'876.23'
# }]
lightCoords = [{
    'x':'12.975301',
    'y':'77.614381'
},
{
    'x':'12.976287',
    'y':"77.614700"
},
{
    'x':'12.973727',
    'y':'77.613877'
},
{
    'x':'12.974372',
    'y':'77.611098'
}]


lightData = [{
    'x':820.87,
    'y':629.27
},
{
    'x':858.29,
    'y':740.50
},
{
    'x':767.64,
    'y':447.973
},
{
    'x':471.54,
    'y':517.86
}]



# lng,lat = net.convertXY2LonLat(float(lightData['y']), float(lightData['x']))
id = ['','1','2', '3']
fileNames = []
outputFileNames=[]
for i in id:
    fileNames.append('traffic'+i+'.signal.xml')
    outputFileNames.append('signal_output'+i+'.json')
    

outputFinal = 'signal_output.json'
```

```python
signalFiles = []
for fileName in fileNames:
    with open(fileName) as fd:
        signalFiles.append (xmltodict.parse(fd.read()) )
```

```python
signalsList = []
for signalFile in signalFiles:
    signalsList.append( signalFile['tlsStates']['tlsState'] )
```

```python
output = []
for i in range(0, len(signalsList[0]), 2):
    if i>500:
        break
    signalObjs = []
    for index, signals in enumerate(signalsList):
        
        signal = signals[i]
#         lng,lat = net.convertXY2LonLat(lightData[index]['y'], lightData[index]['x'], True)
        lat = lightCoords[index]['x']
        lng = lightCoords[index]['y']

        obj = {"id": signal['@id'], "lat":lat, 'lng':lng, 'state':signal['@state']}
        signalObjs.append(obj)
        
    info = {'time':signal['@time'],'signals': signalObjs}
    output.append(info)
```

```python
with open(outputFinal, 'w', encoding='utf-8') as f:
    json.dump(output, f, ensure_ascii=False, indent=4)
```
